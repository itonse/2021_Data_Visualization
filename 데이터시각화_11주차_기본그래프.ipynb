# 면적 그래프(area), 파이 그래프(pie), 고밀도 산점도 그래프(hexbin) 실습

## 강의자료 3p참조
## 기존 코드 그대로 사용하고, 마지막에 df.plot() 안에만 변경하기

```python
import pandas as pd

# 외부에서 CSV 파일 불러오기
data = 'systolic.csv'
df = pd.read_csv(data)

# 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/NGULIM.ttf'  # 폰트가 설치되어 있는 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

new_df = df.iloc[[0, 2, 4, 5, 7], 8:]   
new_df.index = ['이겨레', '강산애', '오대양', '송가인', '황산성']  
print(new_df.head())      # 선택된 데이터셋 출력
new_df.plot(kind='area')   # 면적 그래프 (각각의 값은 y축 높이이다. 이겨레의 1~5차 높이 합: 누적 높이 y값)
new_df.plot(kind='pie', y='1차')   # 파이 그래프 그리기   # y 값은 한 컬럼만 지정 가능 (여기선 1차인 경우만 그래프로 그림)
new_df.plot(kind='hexbin', x='1차', y='2차', gridsize=20)   # 고밀도 산점도 그래프 그리기, x, y에 컬럼 하나씩 지정,  gridsize 설정필요
                                                             # 각 점의 위치는  이름(1차값, 2차값) 이다. ex) 송가인(143, 152)
                                                              # 점 모양이 육각형이라 hex 
```

# 히스토그램(hist)  
## 기존코드, 마지막df.plot()만 변경

```python
import pandas as pd

# 외부에서 CSV 파일 불러오기
data = 'systolic.csv'
df = pd.read_csv(data)

# 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/NGULIM.ttf'  # 폰트가 설치되어 있는 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 행은 인덱스로 2명을 선택할 원소를 정의하고 열은 인덱스 8부터 끝가지 범위 설정
new_df = df.iloc[[4, 5], 8:]   # 2행과 5열의 원소를 인덱스로 설정  
new_df.index = ['오대양', '송가인']  # 2명에 대한 열 이름 지정

print(new_df.head())      # 선택된 데이터셋 출력
new_df.plot(kind='hist', bins=20)   # 히스토그램 그리기   # 막대가 서로 겹치지 않게 bins로 통해 막대의 폭을 좁게 지정 (bins=10 해보기)
```

## 오대양 한 사람의 히스토그램

```python
import pandas as pd

# 외부에서 CSV 파일 불러오기
data = 'systolic.csv'
df = pd.read_csv(data)

# 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/NGULIM.ttf'  # 폰트가 설치되어 있는 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 행은 인덱스로 한명을 선택할 원소를 정의하고 열은 인덱스 8부터 끝가지 범위 설정
new_df = df.iloc[[4], 8:]   # 1행과 5열의 원소를 인덱스로 설정  
new_df.index = ['오대양']  # 한명에 대한 열 이름 지정

print(new_df.head())      # 선택된 데이터셋 출력
new_df.plot(kind='hist', alpha=0.5)   # 히스토그램 그리기 겹쳐서 1차 막대가 4차 막대에 가려지는데, alpha(투명도)를 이용하면 색상이 혼합됨
```
