# 22p 도전문제 (코드 안보고 직접 짤수 있어야 함)

```python
import pandas as pd
import matplotlib.pyplot as plt 

data = 'systolic.csv'
df = pd.read_csv(data)

# 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/NGULIM.ttf'  # 폰트가 설치되어 있는 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)


plt.figure(figsize=(8, 5))

x1 = df['이름']
y1 = df['키']
plt.plot(x1, y1, '-.sm', markersize=10, markerfacecolor=[0.8, 0.5, 0.8])

x2 = df['이름']
y2 = df['체중']
plt.bar(x2, y2, color ='y')

plt.title('연령별 키와 체중 수치 현황', fontsize=15)  
plt.xlabel('참가자 이름')       
plt.ylabel('키와 체중')         
plt.ylim(10, 300)

plt.legend(['키', '체중'])
plt.savefig('연령별 키와 체중 수치 현황.jpeg')
plt.show()
```

## lmplot (선형 회귀 시각화)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 주피터 노트북을 실행한 브라우저에 그래프가 나타나도록 설정
# %matplotlib inline

# 데이터 파일 불러오기
health = 'systolic.csv'
df = pd.read_csv(health)
print(df.head())

# 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/NGULIM.ttf'  # 폰트가 설치되어 있는 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 연령과 BMI 관계의 선형 회귀 시각화 (hue=색상을 적용할 대상 열)
sns.lmplot(x='연령', y='BMI', data=df, hue='성별')
plt.xlim(15, 80)   # x축 범위값 15~80으로 설정
plt.ylim(15, 30)   # y축 범위값 15~30으로 설정

# 그래프 관련 출력 메시지 출력 생략 
plt.show()  
```

## barplot (막대그래프) 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 주피터 노트북을 실행한 브라우저에 그래프가 나타나도록 설정
%matplotlib inline  

health = 'systolic.csv'
df = pd.read_csv(health)
print(df.head())

# 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/NGULIM.ttf'  # 폰트가 설치되어 있는 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 그래프의 크기 설정(단위는 인치)
plt.figure(figsize=(10, 5)) 

# 바 플롯 그리기 (열 제외: 환자코드, 성별, 혈액형)
new_df = df.drop(['환자코드', '성별', '혈액형'], axis=1)  # axis=0: 행에 대해서, axis=1: 열에 대해서
sns.barplot(data=new_df)        # 바 플롯
plt.title('연령별 생체 데이터 분석 현황', fontsize=17)  # 그래프 제목 설정

# 그래프 관련 출력 메시지 출력 생략 
plt.show() 
```

# 서브 플롯 시각화

https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

health = 'systolic.csv'
df = pd.read_csv(health)
print(df)

# 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/NGULIM.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

plt.figure(figsize=(10, 5)) 

# 한 행에 그래프를 4개씩 배치한 서브 플롯 그리기 
g = sns.catplot(data=df, kind='bar', col='BMI', col_wrap=4)   # kind='bar': barplot 그래프, BMI 대한 그래프를 그리겠다(오름차순), 한 행에 4개 그래프 그림 
g.fig.set_size_inches(20, 10)   # 서브 플롯의 크기 설정(가로 20, 세로 10)
g.fig.subplots_adjust(wspace=0.3, hspace=0.3)   # 간격 설정

plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt


health = 'systolic.csv'
df = pd.read_csv(health)

new_df = df.drop(['환자코드', '성별', '연령', '혈액형', '키', '체중'], axis=1)  
print(new_df)

# 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/NGULIM.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 한 행에 그래프를 3개씩 배치한 서브 플롯 그리기 
g = sns.catplot(data=new_df, kind='box', col='5차', col_wrap=3, legend=True)
g.fig.set_size_inches(15, 15)   # 서브 플롯의 가로와 세로 크기 설정
g.fig.subplots_adjust(wspace=0.1, hspace=0.1)  # 간격 설정
plt.ylim(100, 180)             # y축 범위값 100~180으로 설정

plt.ylim(100,180)

plt.show()
```
