```python
import pandas as pd
import matplotlib.pyplot as plt    # 데이터를 차트화 시켜주는 라이브러리

data = 'systolic.csv'
df = pd.read_csv(data)

# 한글 폰트 오류 해결
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/NGULIM.ttf'  # 폰트가 설치되어 있는 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 그래프의 크기 설정(단위는 인치)
plt.figure(figsize=(8, 5)) 

# 첫 번째 선 그래프 그리기
# 옵션 - 파선, 15 크기의 별모양 마커, 마커 색상: RGB 색상 지정
x1 = df['이름']
y1 = df['연령']
plt.plot(x1, y1, '--*m', markersize=15, markerfacecolor=[0.1, 0.6, 0.6])    # 마커 색상= 청록색
## plt.plot(x1, y1, '--*k', markersize=15, markerfacecolor=[0, 0, 0])    # 마커 색상= 검은색

# 두 번째 선 그래프 그리기
# 옵션 - 빨간 색상의 일점 쇄선, 13 크기의 원모양 마커(테두리: 빨간색, 채우기: 노란색)
x2 = df['이름']
y2 = df['BMI']
plt.plot(x2, y2, '-.or', markersize=13, markeredgecolor='r', markerfacecolor='y') 
##plt.plot(x2, y2, 'd', markersize=13, markeredgecolor='r', markerfacecolor='y')  # 이렇게 세번째 항목을 변경하면 어떻게 출력되는지 비교해보기
##plt.plot(x2, y2, ':pg', markersize=13, markeredgecolor='r', markerfacecolor='y')   # 선은 점선, 점 모양펜타곤, 선색상은 그린
##plt.plot(x2, y2, ':^g', markersize=13, markeredgecolor='r', markerfacecolor='y')   # 마커 모양이 윗쪽방향 삼각형
##plt.plot(x2, y2, '-.or', markersize=13, markeredgecolor=[0.5, 0.5, 0.5], markerfacecolor='y') # 마커의 엣지컬러: 그레이


# 그래프 제목(크기 15)과 축 제목, y축 범위, 범례
plt.title('연령별 BMI 수치 현황', fontsize=15)  
plt.xlabel('참가자 이름')       # x축 제목
plt.ylabel('연령 분포')         # y축 제목
plt.ylim(0, 100)                # y축 범위값 0~80으로 설정

# 범례 추가 (범례 설명 문자열)
plt.legend(['연령', 'BMI'])

# 그래프를 jpeg 그림 파일로 저장
plt.savefig('연령별 BMI 수치 현황.jpeg') 

# 그래프 관련 출력 메시지 출력 생략 
plt.show()               
```
