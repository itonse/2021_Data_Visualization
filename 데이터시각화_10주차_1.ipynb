# Week09 강의자료 처음부터  매칭해서 보기(이를 바탕으로 실습한 것)

```python
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams["font.size"] = 12
plt.rcParams['xtick.labelsize'] = 12.
plt.rcParams['ytick.labelsize'] = 12.
plt.rcParams['axes.unicode_minus'] = False

CHART_NAME = 'seabornTest_exam'
cnt, PNG, UNDERBAR = 0, '.png', '_'
filename = 'C:/Users/create/DV/tips.csv'

tips = pd.read_csv(filename, encoding='utf-8', index_col=0)

print(tips.columns)
print('-'*30)

print(tips.head(10))    # 앞 10개 행 출력
print('-'*30)

print(tips.describe())    # 요약 통계치 출력(total_bill, tip, size만 수치형 변수라서 이 컬럼들의 통계치만 출력 됨)
print('-'*30)
```

## 산점도와 Rugplot 함수 동시에 그리기

```python
import seaborn as sns

total_bill = tips['total_bill'] # Series
print('total_bill.unique()')     
print(total_bill.unique())   # 토탈 빌 컬럼의 유니크 값만 출력

import numpy as np
print(np.max(total_bill.unique()))   # 최댓값
print(np.min(total_bill.unique()))    # 최솟값

ax = plt.subplots()  # 산점도와 Rugplot 함수 동시에 그리기
sns.scatterplot(data=tips, x='total_bill', y='tip')    # x축은 토탈빌, y축은 팁 컬럼
ax = sns.rugplot(data=tips, x='total_bill', y='tip')     # x, y 같은 축 이용해서 rugplot 그리기
ax.set_title('산점도와 Rug Plot')
ax.set_xlabel('총 지급액')
ax.set_ylabel('팁(tip)')
```

## 총 지급액 히스토그램

```python
ax = plt.subplots()
ax = sns.histplot(data=tips, x='total_bill', kde=True, bins=30) # histplot(히스토그램), y라벨은 없음, kde는 커널밀도 곡선, bins는 막대의 개수
ax.set_title('총 지급액 히스토그램')
ax.set_xlabel('총 지급액')
ax.set_ylabel('')
```

## 막대그래프

```python
print("tips['day'].unique()")
print(tips['day'].unique())    # day 컬럼의 유니크 값 출력

label_dict = {'Sun':'일요일', 'Sat':'토요일', 'Thur':'목요일', 'Fri':'금요일'}   # 사전구조 정의

def setLabel(x):    # 함수 정의
    return label_dict[x]    # 해당되는 요일 리턴

idx = 0 # 색상 구분을 위한 카운터 변수
tips['day'] = tips['day'].apply(setLabel)   # 함수를 적용해서 day 컬럼을 영어->한글로 바꿈

print('요일별 갯수')
result = tips.groupby('day')['day'].count()    # 요일별 개수 
print(result)

print("tips['time'].unique()")    # Dinner, Lunch
print(tips['time'].unique())   

print('주간/야간 구분 개수')
result = tips.groupby('time')['time'].count()    # 주간, 야간을 구분하기 위해 팁 데이터셋에 타임별로 그룹핑한 다음 그 개수를 카운팅 함
print(result)

plt.figure()
ax = sns.barplot(x='day', y='tip', hue='time', data=tips, errwidth=0)  # barplot(막대그래프), x축:day, y축:tip의 액수로,
ax.set_title("요일에 따른 팁(tip))")                                    # hue를 time으로 함(Dinner, Lunch 별로), 에러바는 설정x 
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
```

# 박스플롯

```python
print('요일에 따른 따른 팁(tip)의 통계치 정보')
mygrouping = tips.groupby('day')['tip']    # tip 데이터셋을 요일별로 그룹핑한 다음, tip에대한 통계치 정보을 mygrouping에 담고
print(mygrouping.describe())    # 요약 통계치 출력   25%: Q1
 
plt.figure()
ax = sns.boxplot(x='day', y='tip', hue='time', data=tips)     # hue 옵션:time
ax.set_title('요일에 따른 따른 팁(tip)(boxplot)')
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
```

# 바이올린 플롯

```python
plt.figure()
ax = sns.violinplot(x='day', y='tip', hue='time', data=tips)    # violinplot, hue옵션 사용
ax.set_title('요일에 따른 팁(tip)(violinplot)')
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
```

# stripplot

```python
STRIP_SIZE = 3 

plt.figure()
ax = sns.stripplot(x='day', y='tip', hue='time', data=tips, jitter=True, size=STRIP_SIZE)  # stripplot, jitter: 서로 겹치치 않게, size: 점의크기 
ax.set_title('요일에 따른 팁(tip)(stripplot)')
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
plt.legend(loc=1)   # 범례박스의 크기를 상단 오른쪽으로
```

# swarmplot

```python
plt.figure()
ax = sns.swarmplot(x='day', y='tip', hue='time', data=tips)   # swarmplot
ax.set_title("Swarm Plot(hue='time')")
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
plt.legend(loc=1)     # 범례박스 위치
```

# Boxplot과 Strip Plot 같이 그리기

```python
import numpy as np
plt.figure()
sns.boxplot(x='total_bill', y='day', data=tips, whis=np.inf)  # boxplot, x축: 총 지금액, y축: day, whis=np.inf: 아웃라이어를 만들지X
ax = sns.stripplot(x='total_bill', y='day', data=tips, jitter=True, color="0.4")  # stripplot, x축: 토탈 빌, y축: day, color: 0=블랙, 1=흰색
ax.set_title("Boxplot과 Strip Plot로 표현")
ax.set_xlabel('총 지급액')
ax.set_ylabel('요일')
```
